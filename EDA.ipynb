🔹 Definition of Outliers:

Outliers are data points that are significantly different from most of the values in a dataset. They may indicate:

Data entry errors

Measurement variability

Rare or extreme events

Outliers can skew statistical results, so identifying them is essential during EDA.

#✅ **Features Extraction(Q 11)**

🔹 1. Age Group (Convert age into bins)

```python
df['age_group'] = pd.cut(df['age'], bins=[0, 25, 45, 65, 100], labels=['Youth', 'Adult', 'Middle-Aged', 'Senior'])
df
```

✅ Interpretation:

 Why this feature helps:

Simplifies analysis: Converts a continuous numeric column into discrete categories that are easier to analyze and visualize.

Helps identify income patterns by life stage:

Do Seniors tend to earn less?

Are Middle-Aged people the most likely to earn >50K?

Improves model interpretability by giving meaningful context to the age feature.

🔹 2. Is High Earner (Binary label for income)

```python
df['is_high_earner'] = df['income'].apply(lambda x: 1 if x == '>50K' else 0)
df
```

✅ Interpretation:

 Why this is done:

Most ML algorithms (like Random Forest, Logistic Regression, etc.) require numeric labels for classification.

This converts the original income column into a binary target variable, which can now be used for:

Classification modeling

Correlation analysis

Evaluation (accuracy, precision, recall)

🔹 3. Capital Net (Gain minus loss)

```python
df['capital_net'] = df['capital.gain'] - df['capital.loss']
df
```

✅ Interpretation:

 Why this feature matters:

Single-value insight: Rather than analyzing two separate columns (capital.gain, capital.loss), capital_net gives a consolidated view of someone's investment profit or loss.

High net capital values often indicate financial activity or wealth, which can be a strong indicator of high income.

Useful for models that don’t automatically detect patterns across multiple columns — you explicitly define a meaningful interaction.

🔹 4. Workload Category (Categorize work hours)

```python
def workload(hours):
    if hours < 30:
        return 'Part-Time'
    elif hours <= 40:
        return 'Full-Time'
    else:
        return 'Overtime'

df['workload'] = df['hours.per.week'].apply(workload)
df
```

✅ Interpretation:

 Why this feature matters:

Human-readable grouping: Instead of using raw numbers like "38.5 hours," we now have simple, descriptive labels: Part-Time, Full-Time, and Overtime.

Adds semantic meaning:

Overtime workers may correlate with higher income or job stress.

Part-Time workers may earn less due to fewer hours.

Helps with categorical visualizations, such as income distribution across workload levels.
